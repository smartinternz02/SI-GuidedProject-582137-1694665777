```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df
```

```python
df.info()   #information
```

```python
sns.scatterplot(x="total",y="speeding",data=df)
```

Inference --  From the above scatter graph we can see that it is positive weak correlation graph.

```python
sns.scatterplot(x="speeding",y="alcohol",data=df)
```

Inference --  From the above scatter graph we can see that it is positive  correlation graph.

```python
sns.scatterplot(x="alcohol",y="not_distracted",data=df)
```

Inference --  From the above scatter graph we can see that it is positive  correlation graph.

```python
sns.scatterplot(x="not_distracted",y="no_previous",data=df)
```

Inference --  From the above scatter graph we can see that it is positive  correlation graph.

```python
sns.lineplot(x="total",y="speeding",data=df)
```

Inference --  From the above line graph we can see that there is no such particular relation between total and speeding values i.e., sometimes it increases and sometimes it decreases.

```python
sns.lineplot(x="speeding",y="alcohol",data=df)
```

Inference --  From the above line graph we can see that there is no such particular relation between speeding and alcohol values i.e., sometimes it increases and sometimes it decreases.

```python
sns.lineplot(x="alcohol",y="not_distracted",data=df)
```

Inference --  From the above line graph we can see that there is no such particular relation between alcohol and not distracted values i.e., sometimes it increases and sometimes it decreases but most of the time it increases.

```python
sns.lineplot(x="not_distracted",y="no_previous",data=df)
```

Inference --  From the above line graph we can see that there is no such particular relation between not distracted and no previous values i.e., sometimes it increases and sometimes it decreases but at starting first the value drops then it's increasing gradually/

```python
sns.displot(df["speeding"])
```

Inference -- Between 4 to 5 the count value is maximum and in different speeding values the average is around 6.

```python
sns.relplot(x="total",y="speeding",data=df,hue="ins_premium")
```

Inference -- The above graph is a scatter plot with different colors which indicates different values.

```python
plt.subplots(figsize=(20,15))
sns.barplot(data=df,x="total",y="speeding",orient='h')
```

Inference -- The above graph is bar plot which is total vs speeding graph.

```python
plt.subplots(figsize=(20,15))
sns.countplot(x="total",data=df)
```

Inference -- The above count graph is of total vs count and the maximum count is around 2.

```python
sns.jointplot(x="total",y="speeding",data=df)
```

Inference -- The above graph is joint graph which is combination of scatter plot and histogram or bar plot.
